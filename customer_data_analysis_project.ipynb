```python
df.columns = df.columns.str.lower()
df.rename(columns={"purchase amount": "purchase_amount"}, inplace=True)
df.rename(columns={"subscriptionstatus":"subscription_status"},inplace=True)
df.rename(columns={"previouspurchases":"previous_purchases"},inplace=True)
df.rename(columns={"purchase amount (usd)": "purchase_amount"}, inplace=True)
```

```python
labels=["young adult","adult","middle age","senior"]
```

```python
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bia_weekly":14,
    "Annually":365,
    "Every three months":90
}
```

```python
df['purches_frequency_days']= df['frequency of purchases'].map(frequency_mapping)
```
